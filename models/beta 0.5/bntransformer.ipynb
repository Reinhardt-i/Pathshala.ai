```python
from transformers import pipeline

pipe = pipeline("question-answering", model="shawmoon/EkattorBert-multilingual-finetuned-squad_v2", framework='pt')

question = "Where do I live?"
context = "My name is Tim and I live in Sweden."
pipe(question = question, context = context)
# {'score': 0.975547730922699, 'start': 28, 'end': 36, 'answer': ' Sweden.'}
```
